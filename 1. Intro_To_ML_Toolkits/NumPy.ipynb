<h2> <b> Q.1) Create a Numpy array of size 4 x 5. </b> </h2>

```python
# matrix is of size 4 X 5

matrix1 = np.zeros(shape=(4,5))
print(matrix1)
```

<h2> <b> Q.2) Randomly initialize the array. </b> </h2>

```python
# Initializing the matrix1

matrix1 = np.append([[1, 2, 3, 4, 5], [2, 3, 4, 5, 6], [-1, 0, 4, 3, 5]], [[2, 4, -3, 0, 1]], axis=0)
print(matrix1)
```

<h2> <b> Q.3) Get the Transpose of the Matrix that you created. Create a square matrix and find its determinant. </b> </h2>

```python
# transpose of the matrix
transpose = matrix1.T
print(transpose, "\n")

square_matrix = np.zeros(shape=(3,3))
square_matrix = np.append([[1, 0, 1], [9, 5, 2]], [[-1, 7, 4]], axis=0)
print(square_matrix)

# determinant of the square matrix
det = np.linalg.det(square_matrix)
print("\nDeterminant of square matrix is: ", int(det))
```

<h2> <b> Q.4) Create another matrix of size 5 x 4 and randomly initialize it. </b> </h2>
