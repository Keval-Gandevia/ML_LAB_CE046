<h3> Linear Regression using Tensorflow </h3>
<br>

<br/><h2> <b> Import necessary libraries. </b> </h2><br/>

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

<br/><h2> <b> Declare input features and targets. </b> </h2><br/>

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37],[69, 96, 70], [73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], 
                   [69,96, 70], [73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], [69, 96, 70]], dtype='float32')

# Targets (apples, oranges)
targets = np.array([[56, 70], [81, 101], [119, 133], [22, 37], [103, 119], [56, 70], [81, 101], [119, 133], [22, 37], [103, 119], 
                    [56, 70], [81, 101], [119, 133], [22, 37], [103, 119]],dtype='float32')
```

<br/><h2> <b> Split the dataset into training and testing data. </b> </h2><br/>

```python
from sklearn.model_selection import train_test_split

inputs_train, inputs_test, targets_train, targets_test = train_test_split(inputs, targets, test_size = 0.25, random_state = 46)
```

<br/><h2> <b> Define the linear regression model using tensorflow. </b> </h2><br/>

```python
from tensorflow import keras
from tensorflow.keras import layers

model = tf.keras.Sequential([layers.Dense(2,input_shape=(3,))])
```

<br/><h2> <b> Declare number of epochs and learning rate </b> </h2><br/>

```python
EPOCHS = 300
LR = 0.005
```

<br/><h2> <b> Train the data into model </b> </h2><br/>

```python
model.compile(optimizer = tf.optimizers.Adam(learning_rate = LR), loss='mean_absolute_error')
```

```python
model_fit = model.fit(inputs_train, targets_train, epochs=EPOCHS)
```

<br/><h2> <b> Make predictions. </b> </h2><br/>

```python
target_pred = model.predict(inputs_test)

print("Predicted target: ", target_pred)

print("\n\nActual Target: ", targets_test)
```

<br/><h2> <b> Calculate the loss. </b> </h2><br/>

```python
model.evaluate(inputs_test, targets_test)
```
