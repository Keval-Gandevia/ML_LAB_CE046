<h2> <b> Q.1) Draw Scatter Plot between SepalLengthCm and SepalWidthCm for "Iris.csv" file with proper labelling. </b> </h2>

```python
# scatter plot

plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'])
plt.title("Scatter plot between SepalLengthCm and SepalWidthCm")
plt.ylabel("SepalWidthCm")
plt.xlabel("SepalLengthCm")
plt.show()
```

<h2> <b> Q.2) Draw Histogram of SepalLengthCm with proper labelling. </b> </h2>

```python
# histogram

plt.hist(data['SepalLengthCm'], bins = 10)
plt.title("Histogram of SepalLengthCm")
plt.xlabel("SepalLengthCm")
plt.show()
```

<h2> <b> Q.3) Plot bar chart of Species. </b> </h2>

```python
df = pd.DataFrame(data)

X = df.iloc[:, 5]

value = X.value_counts()

temp_list = [];
for i in value:
  temp_list.append(i)

# print(temp_list)

plt.bar(value.keys(), temp_list, color='red')
plt.title("Chart of Species")
plt.ylabel("Name of the species")
plt.xlabel("Number of the species")
plt.show()
```
