# K - Means clustering on 'iris' dataset

<br/><h2> <b> Load the iris data</b> </h2><br/>

```python
from sklearn import datasets

iris = datasets.load_iris()
# iris.target
```

<br/><h2> <b> Extract the features and targets from the data.</b> </h2><br/>

```python
features = iris.data
targets = iris.target
iris.target_names
```

<br/><h2> <b> Apply the K - Means clustering using 2 clusters.</b> </h2><br/>

```python
Kmean = KMeans(n_clusters = 2, random_state = 46)
Kmean.fit(features)
```

<br/><h2> <b> Get the cluster centroids.</b> </h2><br/>

```python
Kmean.cluster_centers_
```

<br/><h2> <b> Get the cluster labels.</b> </h2><br/>

```python
Kmean.labels_
```

```python
!pip install yellowbrick
```

<br/><h2> <b> Visualize the K-means with different number of clusters.</b> </h2><br/>

```python
from yellowbrick.cluster import KElbowVisualizer

# Instantiate a scikit-learn K-Means model
model = KMeans(random_state = 46)

# Instantiate the KElbowVisualizer with the number of clusters and the metric
visualizer = KElbowVisualizer(model, k = (2, 6), metric = 'silhouette', timings = False)

# Fit the data and visualize
visualizer.fit(features)
visualizer.poof()
```

<br/><h2> <b> Calculate silhouette_score.</b> </h2><br/>

```python
from sklearn.metrics import silhouette_score
print(silhouette_score(features, Kmean.labels_))
```

```python
sns.relplot(data = diabetes_features)
```

```python
df = pd.DataFrame(diabetes_features)
df.head()
```
